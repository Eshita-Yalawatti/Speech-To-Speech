Create the biagram dataset (x,y)
Inputs and Targets

```python
import torch

xs, ys = [], [] #two lists x: input, y: target

for n in names[:1]:
  ch = ['.'] + list(n) + ['.']
  for ch1, ch2 in zip(ch, ch[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    print(ch1, ch2)
    xs.append(ix1)
    ys.append(ix2)

#we want tensor and not the list
xs = torch.tensor(xs) #torch.tensor is different from torch.Tensor
ys = torch.tensor(ys)
```

```python
#input neuron cant be fed. So we do one hot encoding

import torch.nn.functional as F
xenc = F.one_hot(xs, num_classes=27).float()
xenc
```

```python
W = torch.randn((27, 1)) #torch.randn fill the tensor with random numbers form normal distribution
xenc @ W #@ is the symbol for matrix multiplication in pytorch
```

```python
g = torch.Generator().manual_seed(214743676)
W = torch.randn((27, 27), generator = g, requires_grad = True)
xenc @ W
```

```python
#we want probability distribution and hence we exponentiate these numbers on the so that they can get closer to the count of biagram frequency
#this will convert the negative numbers to less than 1 and positive numbers to more than 1
(xenc @ W).exp()
```

```python
logits = xenc @ W #log-counts
counts = logits.exp() #equivalent to N
probs = counts / counts.sum(1, keepdims=True)
#the above two lines are called softmax and that takes the logits, exponentiates it and normalizes it
#takes outputs and gives out probability distribution
probs
```

```python
#we are doing a classification job and hence we will use negative likelihood as a loss calculating method
probs[0, 1], probs[1, 1], probs[2, 2], probs[3,9], probs[4, 4], probs[5, 0]   #ys=1, 1, 2, 9, 4, 0
```

```python
#better way of writting it
torch.arange(6)
```

```python
probs[torch.arange(6), ys]
```

```python
#calculating the negative likelihood (loss) of the models prediction
#loss is the average log livelihood
loss = -probs[torch.arange(6), ys].log().mean()
loss
```

```python
#backward pass
W.grad = None #setting it to zero
loss.backward()
```
