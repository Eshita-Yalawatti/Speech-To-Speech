```python
import re

# Function to clean names by removing special characters
def clean_names(file_path, output_path):
    with open(file_path, "r", encoding="utf-8") as file:
        names = file.readlines()

    # Remove non-alphabetic characters from each name
    cleaned_names = [re.sub(r'[^a-zA-Z]', '', name).strip() for name in names if re.sub(r'[^a-zA-Z]', '', name).strip()]

    # Save cleaned names to a new file
    with open(output_path, "w", encoding="utf-8") as cleaned_file:
        cleaned_file.write("\n".join(cleaned_names))

    print(f"Cleaned names saved to: {output_path}")

# Example usage
input_file = "Indian_Names.txt"  # Replace with your actual input file path
output_file = "Indian_Names_Cleaned.txt"  # Replace with desired output file path
clean_names(input_file, output_file)
```

```python
import re

# Function to clean names by removing special characters and spaces
def clean_names(file_path, output_path):
    with open(file_path, "r", encoding="utf-8") as file:
        names = file.readlines()

    # Remove non-alphabetic characters and spaces from each name
    cleaned_names = [re.sub(r'[^a-zA-Z]', '', name) for name in names if re.sub(r'[^a-zA-Z]', '', name)]

    # Save cleaned names to a new file
    with open(output_path, "w", encoding="utf-8") as cleaned_file:
        cleaned_file.write("\n".join(cleaned_names))

    print(f"Cleaned names saved to: {output_path}")

# Example usage
input_file = "/content/Indian_Names_Cleaned.txt"  # Replace with your actual input file path
output_file = "/content/Indian_Names_Hi_C.txt"  # Replace with desired output file path
clean_names(input_file, output_file)
```

```python
namesc = open('Indian_Names_Hi_C.txt' , 'r').read().splitlines()
```

```python
import random
random.shuffle(namesc)
```

```python
NC = torch.zeros((27, 27), dtype=torch.int32) #actual array for the dataset
```

```python
charNC = sorted(list(set(''.join(namesc))))
stoiNS = {s:i+1 for i,s in enumerate(charNC)}
stoiNS['.'] = 0
itosNS = {i:s for s,i in stoiNS.items()}
```

```python
stoiNS
```

```python
for n in namesc:
  chNC = ['.'] + list(n) + ['.']
  for ch1, ch2 in zip(chNC, chNC[1:]):
    ix1 = stoiNS[ch1]
    ix2 = stoiNS[ch2]
    NC[ix1, ix2] += 1
```

```python
chNC
```

```python
plt.figure(figsize=(16, 16))
plt.imshow(NC, cmap='Blues')
for i in range(27):
  for j in range(27):
    chrNS = itosNS[i] + itosNS[j]
    plt.text(j, i, chrNS, ha="center", va="bottom", color="gray")
    plt.text(j, i, N[i,j].item(), ha="center", va="top", color='gray')
plt.axis('off');
```
